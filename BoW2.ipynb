```python
import nltk
```

```python
from collections import Counter
```

```python
# sample corpus
corpus = [
    'I love machine learning',
    'Machine learning is fun',
    'I love coding'
]
```

```python
# Build vocabulary
vocab = set()
for sentence in corpus:
    for word in sentence.lower():
        vocab.add(word)

vocab = sorted(vocab)
```

```python
# create BoW Vectors
bow = []
for sentence in corpus:
    word_count = Counter(sentence.lower().split())
    vector = [word_count[word] for word in vocab]
    bow.append(vector)
```

```python
print("Vocabulary:")
print(vocab)
```

```python
print("\nBoW Matrix:")
for row in bow:
    print(row)
```
