```python
# Vocabulary
vocab = set()
for sentence in corpus:
    vocab.update(sentence.split())
vocab = sorted(vocab)
```

```python
# Term Frequency (TF)
tf = []
for sentence in corpus:
    word_count = Counter(sentence.split())
    total_words = len(sentence.split())
    tf.append([word_count[word] / total_words for word in vocab])
```
